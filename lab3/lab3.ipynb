```python
# ad 10 jakie kategorie filmow zostaly wypozyczone przez klientow?

rental = pd.read_sql('SELECT DISTINCT inventory_id FROM rental',con=connection_sqlalchemy)
inventory = pd.read_sql('SELECT film_id, inventory_id FROM inventory',con=connection_sqlalchemy)
film_category = pd.read_sql('SELECT film_id, category_id FROM film_category',con=connection_sqlalchemy)
category = pd.read_sql('SELECT name, category_id FROM category',con=connection_sqlalchemy)

rented_films = inventory[inventory.inventory_id.isin(rental.inventory_id.tolist())]
rented_films = rented_films.drop_duplicates(subset=['film_id'],ignore_index=True)
rented_films_id = rented_films.film_id.tolist()

rented_film_category = film_category[film_category.film_id.isin(rented_films_id)]
rented_film_category = rented_film_category.drop_duplicates(subset=['category_id'],ignore_index=True) 
rented_category_id = rented_film_category.category_id.tolist()

rented_category = category[category.category_id.isin(rented_category_id)]
rented_category = rented_category.drop_duplicates(subset=['category_id'],ignore_index=True) 

rented_category
#category
```
