```python
#zadanie 1
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt


df = pd.read_csv('SCORES.csv', sep = ',')
```

```python
#Wstepny wykres uzyskanych danych
plt.scatter(df['Hours'],df['Scores'])
plt.grid()
plt.title("Wynik uzyskany na tescie w porownaniu do czasu poswieconego na nauke")
plt.xlabel('Ilosc poswieconych godzin na nauke')
plt.ylabel('Wynik uzyskany na tescie')
plt.show()
```

```python
matrix = df.to_numpy()
matrix
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(matrix[:,0],matrix[:,1],train_size=0.3)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train.reshape(-1,1),y_train.reshape(-1,1))
reg.coef_
```

```python
y_predicted = reg.predict(x_test.reshape(-1,1)) #porownujemy teraz z y_test
y_predicted
```

```python
y_test.reshape(-1,1)
```

```python
#ad 8
x_line = np.linspace(1,10)
y_line = reg.coef_[0]*x_line+reg.intercept_[0]
fig,axs = plt.subplots(2)
axs[0].scatter(df['Hours'],df['Scores'])
axs[0].title.set_text("Wykres oryginalny")
axs[1].scatter(x_test,y_predicted)
axs[1].title.set_text("Wartosci przewidziane przez liniowa regresje")
axs[1].plot(x_line,y_line,'r')
```

```python
#ad9
from sklearn.metrics import mean_absolute_error, mean_squared_error
mean_absolute_error(y_test,y_predicted) #sredni blad bezwgledny
```

```python
mean_squared_error(y_test,y_predicted) #blad sredniokwadratowy
```

```python
np.sqrt(mean_squared_error(y_test,y_predicted)) #pierwiastek bledu sredniokwadratowego
```

```python
#zadanie 2
from sklearn.preprocessing import StandardScaler


df3 = pd.read_csv('PETROL.csv')
df3
```

```python
X = df3.iloc[:, :-1].values
y = df3.iloc[:, -1].values

sc = StandardScaler()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
```

```python
mean_absolute_error(y_pred, y_test)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
#zadanie 3
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

df4 = pd.read_csv('HEART.csv')

df4 = df4.drop(['slope', 'ca', 'thal'], axis=1)

df4 = df4.replace('?', 0)

df4 = pd.get_dummies(df4, columns=['cp', 'restecg'])
df4
```

```python
X = df4.iloc[:, :-1].values
y = df4.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

regressor = LogisticRegression()

regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

print(classification_report(y_test, y_pred))

print('Confusion matrix')
print(confusion_matrix(y_test, y_pred)) #bez printa dzieja sie jakies dziwne rzeczy
```
